# Object-oriented neural network

```python
from sklearn.datasets import make_moons
import numpy as np
import matplotlib.pyplot as plt
from utils.activations import ReLU, Sigmoid
from utils.losses import NLL
from model import NeuralNetwork
```

### Generate dataset

```python
X, Y = make_moons(noise=0.1)
Y = np.expand_dims(Y, 1)
plt.style.use('dark_background')
plt.scatter(X.T[0], X.T[1], c=Y, edgecolors='black')
plt.show()
```

### Define model

```python
model = NeuralNetwork([
    {'in_channels': X.shape[1], 'out_channels': 4, 'activation': ReLU()},
    {'in_channels': 4, 'out_channels': 4, 'activation': ReLU()},
    {'in_channels': 4, 'out_channels': Y.shape[1], 'activation': Sigmoid()}
], NLL())
```

### Training

```python
n_iterations = 2000
alpha = 0.1

model.train(X, Y, n_iterations, alpha)
```
